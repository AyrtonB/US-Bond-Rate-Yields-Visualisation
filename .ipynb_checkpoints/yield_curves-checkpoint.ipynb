```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
from matplotlib import cm
import seaborn as sns

import requests
import xmltodict

from datetime import datetime

import io
import base64
import imageio
from IPython.display import Image, HTML
import moviepy.editor as mp
```

```python
## Website = https://www.treasury.gov/resource-center/data-chart-center/interest-rates/pages/TextView.aspx?data=yieldAll
## XML = https://data.treasury.gov/feed.svc/DailyTreasuryYieldCurveRateData

xml_url = 'https://data.treasury.gov/feed.svc/DailyTreasuryYieldCurveRateData'

response = requests.get(xml_url)
xml_r = xmltodict.parse(response.text)
```

```python
df_yields_raw = pd.DataFrame()

for entry_dict in xml_r['feed']['entry']:
    entry_dict_df = pd.DataFrame(entry_dict['content']['m:properties']).loc['#text']
    df_yields_raw = df_yields_raw.append(entry_dict_df)
    
df_yields_raw.head()
```

```python
df_yields = (df_yields_raw
             .copy()
             .reset_index(drop=True)
             .drop(columns=['d:Id', 'd:BC_30YEARDISPLAY'])
            )

new_col_names = [col.split('_')[1].lower() for col in df_yields.columns]
col_rename_map = dict(zip(df_yields.columns, new_col_names))
df_yields = df_yields.rename(columns=col_rename_map)

df_yields = (df_yields
            .assign(date=pd.to_datetime(df_yields['date']))
            .set_index('date')
            .sort_index()
            )

for col in df_yields.columns:
    df_yields[col] = df_yields[col].astype(float)

df_yields.head()
```

```python
df_yield_cols = (pd.DataFrame(df_yields.columns, columns=['original_string'])
                 ['original_string']
                 .str.split('(\d+)([A-z]+)', expand=True)
                 [[1, 2]]
                 .rename(columns={1:'value', 2:'unit'})
                 .assign(col_names=df_yields.columns)
                )

df_yield_cols.head()
```

```python
df_yield_cols['value'] = df_yield_cols['value'].astype(int)
df_yield_cols['month_value'] = df_yield_cols['value']
s_year_replacements = df_yield_cols[df_yield_cols['unit']=='year']['value']*12
df_yield_cols.loc[s_year_replacements.index, 'month_value'] = s_year_replacements
df_yield_cols = df_yield_cols.sort_values('month_value')

df_yield_cols.head()
```

```python
df_yields = df_yields[df_yield_cols.col_names.values]
    
df_yields.head()
```

```python
fig = plt.figure(dpi=150)
ax = plt.subplot()

sns.heatmap(df_yields.iloc[:, 2:-2], ax=ax)
```

```python
bond_periods = df_yields.columns[2:-3]
cmap = cm.get_cmap('viridis', len(bond_periods))
cmap_dict = dict(zip(bond_periods, cmap.colors))

## Plotting
fig = plt.figure(dpi=150)
ax = plt.subplot()

ax.plot([df_yields.index.min(), df_yields.index.max()], [1, 1], 'k--', label='Parity')

for bond_period in bond_periods:
    (df_yields[bond_period]/df_yields['10year']).plot(label=f'{bond_period[0]} {bond_period[1:].capitalize()}', 
                                                      color=tuple(cmap_dict[bond_period][:3]), ax=ax)

ax.plot([date(2001, 3, 1), date(2001, 3, 1)], [0, 1], color='yellow')
ax.plot([date(2001, 3, 1), date(2001, 3, 1)], [0, 1], 'k:')
ax.plot([date(2007, 8, 1), date(2007, 8, 1)], [0, 1], color='yellow')
ax.plot([date(2007, 8, 1), date(2007, 8, 1)], [0, 1], 'k:')

plt.ylim(0, 1.3)
plt.xlim(df_yields.index.min(), df_yields.index.max())

plt.title('US Treasury Yield Rates')
plt.ylabel('Ratio to 10 Year Yield Rate')
plt.xlabel('')

plt.legend(frameon=False, bbox_to_anchor=(1.05, 0.85))

ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
```

### Creating GIF

```python
def plot_yield_rate_ratios(df_yields, upper_date):
    if isinstance(upper_date, pd.Timestamp):
        upper_date = upper_date.to_pydatetime()
        
    df_yields_plot = df_yields[:upper_date]
    
    ## Plotting
    fig = plt.figure(dpi=200)
    ax = plt.subplot()

    ax.plot([df_yields.index.min(), df_yields.index.max()], [1, 1], 'k--', label='Parity')

    for bond_period in bond_periods:
        ((df_yields_plot[bond_period]/df_yields_plot['10year'])
         .plot(label=f'{bond_period[0]} {bond_period[1:].capitalize()}', 
              color=tuple(cmap_dict[bond_period][:3]), ax=ax)
        )

    recession_2001 = datetime(2001, 3, 1)
    if upper_date > recession_2001:
        ax.plot([recession_2001, recession_2001], [0, 1], color='yellow')
        ax.plot([recession_2001, recession_2001], [0, 1], 'k:')
        
    recession_2007 = datetime(2007, 8, 1)
    if upper_date > recession_2007:
        ax.plot([recession_2007, recession_2007], [0, 1], color='yellow')
        ax.plot([recession_2007, recession_2007], [0, 1], 'k:')

    plt.ylim(0, 1.3)
    plt.xlim(df_yields.index.min(), df_yields.index.max())

    plt.title('US Treasury Yield Rates')
    plt.ylabel('Ratio to 10 Year Yield Rate')
    plt.xlabel('')

    plt.legend(frameon=False, bbox_to_anchor=(1.05, 0.85))

    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
    
plot_yield_rate_ratios(df_yields, datetime(2006, 1, 1))
```

```python
upper_dates = pd.date_range(df_yields.index.min(), df_yields.index.max(), freq='M')

for upper_date in upper_dates:
    plot_yield_rate_ratios(df_yields, upper_date)
    plt.savefig(f'img/{upper_date.value}.png', bbox_inches='tight')
    plt.close()
```

```python
gif_n_mp4_plots = True

if gif_n_mp4_plots == True:
    images = []
    for upper_date in upper_dates:
        filepath = f'img/{upper_date.value}.png'
        
        images.append(imageio.imread(filepath))

        if upper_date == upper_dates[-1]:
            for i in range(3):
                images.append(imageio.imread(filepath))

    kwargs = {'duration': 0.5}
    imageio.mimsave('img/US_bond_yields.gif', images, 'GIF', **kwargs)
    clip = mp.VideoFileClip('img/US_bond_yields.gif')
    clip.write_videofile('img/US_bond_yields.mp4')
```
